```python
movies = pd.read_csv("movies.csv")
movies
```

```python
def convert_to_imdb_link(imdbid):
    return f"https://www.imdb.com/title/tt{'0' * (7 - len(str(int(imdbid))))}{imdbid}/"
```

```python
def convert_to_tmdb_link(tmdbid):
    if tmdbid == "0":
        return str("Sorry :( , Link is not available ! ")
    else :
        return str(f"https://www.themoviedb.org/movie/{str(int(tmdbid))}/")
```

```python
imdb_link = links["imdbId"].apply(convert_to_imdb_link)
tmdb_link = links["tmdbId"].apply(convert_to_tmdb_link)
```

# `Extract genres separated : `

```python
def extract_all_genres(genres_series):
    genre_set = set()
    for genres in genres_series:
        genres_list = genres.split("|")
        genre_set.update(genres_list)
    return genre_set
```

```python
genres_set = extract_all_genres(films_metadata["genres"][:])
```

```python
genres_set
```

```python
len(genres_set)
```

```python
for genre in genres_set:
    print(genre)
```

# `One Hot Encodeing genres : `

```python
def one_hot_encoding(genres_series):
    all_genres = extract_all_genres(genres_series)
    genre_dict = {genre: i for i, genre in enumerate(all_genres)}
    encoded_genres = []
    for genres in genres_series:
        genres_list = genres.split("|")
        encoding = [0] * len(all_genres)
        for genre in genres_list:
            encoding[genre_dict[genre]] = 1
        encoded_genres.append(encoding)
    return encoded_genres, genre_dict

genres_series = movies["genres"]
encoded_genres, genredict = one_hot_encoding(genres_series)
print(encoded_genres[0])
genredict
```

```python
genres_encoded = pd.DataFrame(columns=genredict.keys())
for i, genre in enumerate(encoded_genres):
    genres_encoded.loc[i] = genre
```

```python
genres_encoded
```

```python
genres_encoded.to_csv("Genres_encoded.csv")
```

# `Read user-item ratings : `

```python
ratings = pd.read_csv("ratings.csv")
ratings
```

# `Averaging rating per user  : `

```python
average_ratings_users = ratings.groupby(['userId'])['rating'].mean().reset_index().sort_values(by="rating")
average_ratings_users
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(round(average_ratings_users["rating"], 2), bins=30, kde=True);
plt.title('Distribution of Ratings of per user')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# `Averaging rating per movies  : `

```python
average_ratings_movie = ratings.groupby(['movieId'])['rating'].mean().reset_index().sort_values(by="rating")
average_ratings_movie
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(round(average_ratings_movie["rating"], 1), bins=20, kde=True);
plt.title('Distribution of Ratings of movies')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# `Films metadata with ratings  : `

```python
merged_data = pd.merge(films_metadata, average_ratings_movie, left_on='movieID', right_on='movieId', how='left')

films_metadata['users_avg_ratings_to_movie'] = merged_data['rating']
```

# `Plot distribution of ratings per genre  : `

```python
def plot_genre_ratings_distribution(df, genre_column='genres', rating_column='users_avg_ratings_to_movie', genres_encoded=None):
    if genres_encoded is None:
        genres_encoded = df.columns.difference([genre_column, rating_column])
    
    for genre in genres_encoded:
        if genre not in df.columns:
            print(f"Genre '{genre}' not found in DataFrame columns.")
            continue
        
        genre_df = df[df[genre] == 1]
        plt.figure(figsize=(12, 8))
        sns.histplot(genre_df[rating_column], kde=True)
        plt.title(f'Distribution of Ratings for {genre}')
        plt.xlabel('Rating')
        plt.ylabel('Frequency')
        plt.show()

```

```python
plot_genre_ratings_distribution(films_metadata_with_genre_enc, genres_encoded=genres_encoded)
```
